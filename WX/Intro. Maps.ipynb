```python
# http://introtopython.org/visualization_earthquakes.html
%matplotlib inline
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
#import matplotlib
#matplotlib.rcParams['figure.figsize'] = [15.0, 12.0]
#import numpy as np

# http://www.latlong.net/
# http://matplotlib.org/basemap/api/basemap_api.html#module-mpl_toolkits.basemap
m = Basemap(projection='merc', lat_0=40.480381, lon_0=-89.208984,
    resolution = 'h', llcrnrlon=-127.089844, llcrnrlat=24.141741,
    urcrnrlon=-65.039063, urcrnrlat=50.467995, area_thresh = 1000)
m.drawcoastlines()
#m.fillcontinents(lake_color='b') 
m.drawstates()
#m.drawcounties()
m.drawcountries()
plt.show()
```
