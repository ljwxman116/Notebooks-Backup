```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pyplot import axvline
import matplotlib
matplotlib.rcParams['figure.figsize'] = [15.0, 12.0]
import numpy as np
```

```python
x=[.55,.86,1.39,1.62]
y=[1.649,2.365,3.695,4.660]
```

```python
# -*- coding: utf-8 -*-
#http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.linregress.html
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
r=r_value**2
print 'R²=%r'% r
print slope
```

```python
m=slope
b=intercept
```

```python
# -*- coding: utf-8 -*-
fig, ax = plt.subplots()
plt.style.use('bmh') 
fit = np.polyfit(x,y,1)
fit_fn = np.poly1d(fit)
#Extrapolated data (line)
x1 = np.linspace(0.5, 1.9)
y1 = m*x1+b
plt.plot(x1, y1,'--k') #--k means black colored dashes

#http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.vlines
plt.vlines(1.25,2.119,3.49,colors='k',linestyles='dotted', linewidth=4)

#http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hlines
plt.hlines(2.119,.75,1.25,colors='k',linestyles='dotted', linewidth=4)

plt.plot(x,y,'bo',x,fit_fn(x),'k')
plt.xlim(.5, 1.8)
plt.ylim(1, 5)
ax.set_title('Mass vs. Volume (Unknown Metal #14)', fontsize=24)
ax.set_xlabel('Volume (mL)', fontsize=18)
ax.set_ylabel('Mass (g)', fontsize=18)
plt.text(1.53, 1.05, r'$R^{2}$=%.3f' % r, fontsize=17) #http://matplotlib.org/users/mathtext.html
plt.text(1.53, 1.2, 'Density=%.2f g/mL' % m, fontsize=17) #https://docs.python.org/2/library/stdtypes.html#string-formatting
plt.text(1, 2.01, r'$\Delta$X' % r, fontsize=17)
plt.text(1.252, 2.75, r'$\Delta$Y' % r, fontsize=17)
plt.text(1.685, .78, 'Lucas Reynolds', fontsize=10)
plt.text(.61, 2.12, '(0.750,2.119)', fontsize=12)
plt.text(1.145, 3.52, '(1.25,3.49)', fontsize=12)
```
